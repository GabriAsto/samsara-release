```python
# load PSRFs data
path = "../data/convergence/psrf_results.txt"
psrf_results = np.loadtxt(path, unpack=True)
```

```python
# extract the max psrf values and the corresponding generations for each psrf computed
psrf = psrf_results[0,:]
Ngen = psrf_results[1,:]
```

```python
# plot trace of the maxPSRF
plt.figure()
plt.plot(Ngen, psrf, color='k', label='$\\texttt{samsara}$')
plt.axhline(1., color='r', linestyle='--', label='reference')
plt.ylabel('$\\max \\hat{R}_v$')
plt.xlabel('Generation')
plt.xlim(0., 1050.)
plt.legend(loc='upper right', fontsize=14.5, frameon=False)
plt.savefig("../plots/convergence/Figure9.pdf", bbox_inches='tight')
plt.show()
```
